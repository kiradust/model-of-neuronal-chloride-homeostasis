```python
from plm_singlecomp_withkcc2 import plm, R, nao, zp
from plotting import minithreefig
import matplotlib.pyplot as plt
import numpy as np
from fig5 import f5d
from figure_2or3 import delta_gs3
# electroneutral 2Na/H2PO4^2-
# electrogenic 2Na/H2PO4-
# phosphate doubling over time -- intracellular Pi is 5-10 mM
```

```python
#dez = f5d()
pi, ena, ek, ecl, exi, ev, nai, ki, cli, xi, vm, w, z = zp(np.linspace(-400,99,1000),graph=1)

header = ['z','vm','na','k','cl','x']
rows = np.array([np.linspace(-4,0.99,1000),np.array(vm),np.array(nai),np.array(ki),np.array(cli),np.array(xi)]).T

import csv
with open('data_fig1.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(rows)
```

```python
#dez = f5d()
pk, cli, ki, nai, w, xi, vm, ena, ek, ecl, ev = delta_gs3(np.linspace(1,1000,1000))

header = ['pk','vm','na','k','cl','x']
rows = np.array([pk,np.array(vm),np.array(nai),np.array(ki),np.array(cli),np.array(xi)]).T

plt.plot(pk,ena,label='ena')
plt.plot(pk,ev,label='vm')
plt.plot(pk,ecl,label='ecl')
plt.plot(pk,ek,label='ek')
plt.legend()
plt.show()

import csv
with open('data_fig2.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(rows)
```
